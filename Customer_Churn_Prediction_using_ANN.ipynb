**Remove Spaces**

```python
# MultipleLines	InternetService	OnlineSecurity	OnlineBackup	DeviceProtection	TechSupport	StreamingTV	StreamingMovies	Contract	PaperlessBilling	PaymentMethod
```

```python
data['MultipleLines'] = data['MultipleLines'].apply(lambda x: x.replace(' ', ''))
data['InternetService'] = data['InternetService'].apply(lambda x: x.replace(' ', ''))
data['OnlineSecurity'] = data['OnlineSecurity'].apply(lambda x: x.replace(' ', ''))
data['OnlineBackup'] = data['OnlineBackup'].apply(lambda x: x.replace(' ', ''))
data['MultipleLines'] = data['DeviceProtection'].apply(lambda x: x.replace(' ', ''))
data['InternetService'] = data['TechSupport'].apply(lambda x: x.replace(' ', ''))
data['OnlineSecurity'] = data['StreamingTV'].apply(lambda x: x.replace(' ', ''))
data['OnlineBackup'] = data['StreamingMovies'].apply(lambda x: x.replace(' ', ''))
data['Contract'] = data['Contract'].apply(lambda x: x.replace(' ', ''))
data['PaperlessBilling'] = data['PaperlessBilling'].apply(lambda x: x.replace(' ', ''))
data['PaymentMethod'] = data['PaymentMethod'].apply(lambda x: x.replace(' ', ''))
```

train Test Split

```python
X=data['tenure']
```

```python
X
```

```python
Y=data['Churn']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

**Relu with Binary Cross Entropy**

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation = 'relu'),
    keras.layers.Dense(50, activation = 'relu'),
    keras.layers.Dense(10, activation = 'relu'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.predict(x_test)
```

**Relu with Mean Square Error**

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation = 'relu'),
    keras.layers.Dense(50, activation = 'relu'),
    keras.layers.Dense(10, activation = 'relu'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.predict(x_test)
```

**Sigmoid with Mean Square Error**

```python
model = keras.Sequential([
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation = 'sigmoid'),
    keras.layers.Dense(50, activation = 'sigmoid'),
    keras.layers.Dense(10, activation = 'sigmoid'),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```
